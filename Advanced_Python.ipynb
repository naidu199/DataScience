```python
#adding all the files
```

```python
dataFrame=[]
for dic in glob.glob("/content/drive/MyDrive/AnalyticsVidhya/DataSets/multi-directory/*"):
  for files in glob.glob(dic+"/*"):
    dataFrame.append(pd.read_csv(files))
```

```python
final_data=pd.concat(dataFrame,axis=0)
```
